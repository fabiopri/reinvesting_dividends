## Introduction

### 1. Why dividends?

Portfolios of **dividend stocks** are very popular among stocks investors in Brazil, where dividends are **tax free** since 1995. There is an ongoing debate about taxing dividends again.

This portfolio strategy requires a lot of **discipline**. There seems always to exist a temptation to use dividend-based income in everyday spendings (which may be a right decision for some investors).

But if you are in the **accumulation period**, i. e., if you are not retired yet, there is a lot of evidence of the **benefits of reinvesting dividends** to make your weatlh grow.

For a **more detailed discussion** of whether to reinvest or not to reinvest dividends, please [check this article](https://www.investopedia.com/articles/investing/090915/reinvesting-dividends-pays-long-run.asp).

### 2. Price return and total return

There are two ways of calculating stocks returns: **price return** and **total return**.

When analysing the performance of a particular stock in any chart or database, **you should always check first if it reflects price return or total return**.

Here is a quick overview:

- **Price return** tracks only capital gains. It is useful for trading strategies based on technical analysis.<p>

- **Total return** tracks capital gains and cash distributions (such as dividends). It is useful for long run strategies and reflects more accurately how much money the company made for investors.

### 3. How much difference does it make to reinvest dividends?

Let's check the difference between price return and total return of **Banco do Brasil (BBAS3)**, a popular dividend stock traded at [**B3**](https://www.b3.com.br/), the Brazilian stock exchange, over a **5-year period** and a **10-year period**.

### 4. Importing libraries

```python
import pandas as pd
import yfinance as yf
import plotly.graph_objects as go
```

### 5. Defining start dates (5 years and 10 years) and end date of the dataset

```python
start_05y = '2018-08-09'
start_10y = '2013-08-09'
end = '2023-08-09'
```

---

## First analysis: 5-year period

### 1. Retrieving BBAS3 stock data from Yahoo Finance  and checking first rows

```python
bbas3_5y = yf.download('BBAS3.SA', start = start_05y, end = end)[['Close', 'Adj Close']]
```

```python
bbas3_5y.head()
```

### 2. Renaming columns

```python
bbas3_5y.columns = ['Dividends not reinvested', 'Dividends reinvested']
```

```python
bbas3_5y.head()
```

### 3. Normalizing returns for an accurate comparison 

```python
bbas3_5y_norm = (bbas3_5y / bbas3_5y.iloc[0]) - 1
```

```python
bbas3_5y_norm
```

### 4. Plotting

```python
fig = go.Figure()

fig.add_trace(go.Scatter(
    x = bbas3_5y_norm.index, y = bbas3_5y_norm['Dividends reinvested'],
    mode='lines', name='Dividends reinvested'))

fig.add_trace(go.Scatter(
    x = bbas3_5y_norm.index, y = bbas3_5y_norm['Dividends not reinvested'],
    mode='lines', name='Dividends not reinvested'))

fig.update_layout(title='Return of BBAS3 in 5 years (aug/2018 - aug/2023)',
                   xaxis_title='Date',
                   yaxis_title='Return')

fig.show('png')
```

## Second analysis: 10-year period

### 1. Retrieving BBAS3 stock data from Yahoo Finance  and checking first rows

```python
bbas3_10y = yf.download('BBAS3.SA', start = start_10y, end = end)[['Close', 'Adj Close']]
```

```python
bbas3_10y.head()
```

### 2. Renaming columns

```python
bbas3_10y.columns = ['Dividends not reinvested', 'Dividends reinvested']
```

```python
bbas3_10y.head()
```

### 3. Normalizing returns for an accurate comparison 

```python
bbas3_10y_norm = (bbas3_10y / bbas3_10y.iloc[0]) - 1
```

```python
bbas3_10y_norm
```

### 4. Plotting

```python
fig = go.Figure()

fig.add_trace(go.Scatter(
    x = bbas3_10y_norm.index, y = bbas3_10y_norm['Dividends reinvested'],
    mode='lines', name='Dividends reinvested'))

fig.add_trace(go.Scatter(
    x = bbas3_10y_norm.index, y = bbas3_10y_norm['Dividends not reinvested'],
    mode='lines', name='Dividends not reinvested'))

fig.update_layout(title='Return of BBAS3 in 10 years (aug/2013 - aug/2023)',
                   xaxis_title='Date',
                   yaxis_title='Return')

fig.show('png')
```

---

## Conclusion 

### Over a 5-year period, investors of BBAS3 who didn't reinvest dividends had a 39.85% return, and those who did reinvest had a 92.79% return.

```python
bar_5y = (bbas3_5y_norm.iloc[-1] * 100)
bar_5y.rename('5-year return', inplace=True)
bar_5y = bar_5y.to_frame()
bar_5y
```

```python
fig = go.Figure(
    layout=go.Layout(height=500, width=700)
)

fig.add_trace(go.Bar(
    x = bar_5y.index, y = bar_5y['5-year return'],
    name='5-year return',
    text=round(bar_5y['5-year return'], 2)
))

fig.update_layout(title='Return of BBAS3 in 5 years (aug/2018 - aug/2023)',
                   yaxis_title='Return (%)')

fig.show('png')
```

### Over a 10-year period, investors of BBAS3 who didn't reinvest dividends had a 107.92% return, and those who did reinvest had a 296.89% return

```python
bar_10y = (bbas3_10y_norm.iloc[-1] * 100)
bar_10y.rename('10-year return', inplace=True)
bar_10y = bar_10y.to_frame()
bar_10y
```

```python
fig = go.Figure(layout=go.Layout(height=500, width=700))

fig.add_trace(go.Bar(
    x = bar_10y.index, y = bar_10y['10-year return'],
    name='5-year return',
    text=round(bar_10y['10-year return'], 2)

))

fig.update_layout(title='Return of BBAS3 in 10 years (aug/2013 - aug/2023)',
                   yaxis_title='Return (%)')

fig.show('png')
```
